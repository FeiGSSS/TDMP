```python
import matplotlib.pyplot as plt
plt.style.use(['science','no-latex','ieee'])
from src.utils import load
```

```python
def plot(prefix):
    import numpy as np
    mc = load(prefix+"MC.pkl")
    ele = load(prefix+"ELE.pkl")
    mmca = load(prefix+"MMCA.pkl")
    rdmp = load(prefix+"rDMP.pkl")
    
    x = np.linspace(0.5, 1.5, 20)
    fig, axs = plt.subplots(1, 2, figsize=(8,3))

    colors = [(237/256, 107/256,171/256), (240/256, 166/256,71/256), (5/256,52/256,139/256)]
    ls = ["--", ":", "-"]

    ax = axs[0]
    ax.plot(x, mc, "o", markersize=6, mfc="none",label="MC")
    ax.plot(x, ele,  ls[0], c=colors[0], label="ELE")
    ax.plot(x, mmca, ls[1], c=colors[1], label="MMCA")
    ax.plot(x, rdmp, ls[2], c=colors[2], label="rDMP")
    ax.legend()

    ax = axs[1]
    ax.plot(x, np.abs(mc-ele),  ls[0], c=colors[0], label="ELE")
    ax.plot(x, np.abs(mc-mmca), ls[1], c=colors[1], label="MMCA")
    ax.plot(x, np.abs(mc-rdmp), ls[2], c=colors[2], label="rDMP")
    plt.show()
```

```python
prefix = "results/er/N500/k_6/"
plot(prefix)
```
